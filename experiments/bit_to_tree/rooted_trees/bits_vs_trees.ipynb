# Understand theoretical best bit/nodes ratio

Compare number of trees with n nodes to their log

```python
import sys
sys.path.append('../../')

import pandas as pd
import plotly.express as px
import tree_lib.util as util

df = pd.read_csv("num_rooted_trees.csv")

def big_log2(n):
    if n <= 2:
        return 0
    f = lambda x: 2**x
    return util.largest_arg_fitting(f, n, False) 

log2nodes = [big_log2(int(fn)) for fn in df['ntrees']]
df['bits'] = log2nodes # How many bits are needed to represent all the trees

df['nodes per bit'] =  df['nodes'] / df['bits']
fig = px.line(df, x="nodes", y="nodes per bit", title="Nodes per Bit")
fig.show()
```
